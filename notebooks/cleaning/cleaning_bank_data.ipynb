## Cleaning Bank Customer data

```python
import pandas as pd
```

```python
data = pd.read_csv("../../data/Bank_Customer-Churn-Records.csv")
```

```python
data.head()
```

```python
data.shape
```

    Dropping useless features

- Also dropping 'Complain' because of 99% correlation with target variable making it a problem in training.
- Also Satisfaction Score and Point Earned are uniformly distributed and creating noise so dropping them is better.

```python
columns_to_drop = ['RowNumber', 'CustomerId', 'Surname', 'Complain', 'Satisfaction Score', 'Point Earned']
data = data.drop(columns=columns_to_drop)
```

```python
print(data.shape)
data.head()
```

```python
# Check count of duplicate rows
print("Duplicate rows:", data.duplicated().sum())
```

    Saving cleaned data

```python
import os

# 1. Define the destination folder path and filename
# (Relative path from your notebook folder to the data folder)
output_dir = '../../data/cleaned'
output_file_path = os.path.join(output_dir, 'Bank_Customer_Churn_Cleaned.csv')

# 2. Create the directory if it doesn't already exist
os.makedirs(output_dir, exist_ok=True)

# 3. Save the cleaned DataFrame (index=False prevents pandas from writing a duplicate index column)
data.to_csv(output_file_path, index=False)

print(f"Cleaned dataset saved successfully to: {output_file_path}")
```

## Cleaning Telco Customer data

```python
import pandas as pd
```

```python
df = pd.read_excel("../../data/Telco_customer_churn.xlsx")
```

```python
df.head()
```

```python
df.shape
```

    Dropping useless noise features

```python
df = df.drop(columns=['CustomerID', 'Count', 'Country', 'State', 'Churn Label', 'Churn Score', 'CLTV', 'City', 'Zip Code', 
    'Lat Long', 'Latitude', 'Longitude',])
```

```python
print(df.shape)
df.head()
```

    Check duplicate rows

```python
# Check count of duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print("Duplicate rows:", df.duplicated().sum())
```

    Checking null values

```python
df.isnull().sum()
```

    Removing null values

```python
df = df.drop(columns=['Churn Reason']) 
```

```python
df.isnull().sum()
```

    Checking data types

```python
# Print info to verify data types
df.info()
```

    Summary of data

```python
# Inspect summary statistics
df.describe()
```

```python
df.head(10)
```

    Saving cleaned data

```python
import os

# 1. Define the destination folder path and filename
# (Relative path from your notebook folder to the data folder)
output_dir = '../../data/cleaned'
output_file_path = os.path.join(output_dir, 'Telco_customer_churn_cleaned.csv')

# 2. Create the directory if it doesn't already exist
os.makedirs(output_dir, exist_ok=True)

# 3. Save the cleaned DataFrame (index=False prevents pandas from writing a duplicate index column)
df.to_csv(output_file_path, index=False)

print(f"Cleaned dataset saved successfully to: {output_file_path}")
```
